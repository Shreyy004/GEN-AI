```python
from langchain_ollama import OllamaEmbeddings
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
embeddinngs=OllamaEmbeddings(model='nomic-embed-text',base_url="http://localhost:11434")
```

```python
vector=embeddinngs.embed_query("hello world")
```

```python
vector
```

```python
len(vector)
index=faiss.IndexFlatL2(len(vector))
index.ntotal , index.d
```

```python
vector_store=FAISS(
    embedding_function=embeddinngs,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
db_name="health_supplements"
vector_store=FAISS.load_local(db_name,embeddinngs,allow_dangerous_deserialization=True)
```

```python
question="What are the benefits of taking vitamin? "
docs=vector_store.similarity_search(query=question,k=3)
```

```python
docs
```

```python
retriever=vector_store.as_retriever(search_type="similarity",
                                    search_kwargs={'k':3})
```

```python
retriever.invoke(question)
```
